## This notebook briefly shows how to integrate functions using Vegas

```python
import vegas  #pip3 install vegas if not installed
import numpy as np
```

#### 1D integral

```python
# function to be integrated
def f(x):
    x=x[0]  # This is needed because vegas uses lists [x1,x2,...] as input for the functions and we need to extract the first component since it's a 1d integral
    return x**2 

# Integration Domain
integ = vegas.Integrator([[0, 1]])  

# Integrating the function f(x)
result = integ(f, nitn=10, neval=1000)

# Usa .mean e .sdev correttamente per RAvg
print("Result: ", result.mean)
print("Error: ", result.sdev)
print("Expected Result: ", 1/3)


```

### 2D Integral

```python
# Function to be integrated
def integrand(variables):
    x, y = variables # This line "unpacks" the list variables = [x1,x2] assigning x = x1, y = x2 
    return np.exp(-(x**2 + y**2))


integral = vegas.Integrator([[0, 1], [0, 1]])

result = integral(integrand, nitn=10, neval=500)
print(result)
# Mostrare il risultato
print("Integral: ", result.mean)
print("Error: ", result.sdev)
print("Mathematica result: ",0.557746)
```

```python
##Â studiarlo un po meglio ma questa funzione dovrebbe restituire gli eventi montecarlo che ha generato per calcoalre l'integrale che gli ho passato
## il print dei pesi serve ad assicurarsi che questi siano simili e quindi ogni zona dello spazio delle fasi sia stata popolata correttamente

integral_0 = 0.0
count=0

for x, wgt in integral.random():
    integral_0 += wgt * f(x)
    count+=1
print(count)
x_points = np.zeros((count ,2))
w_points = np.array([0.]*count)

i=0
for x, wgt in integral.random():
    x_points[i,0]=x[0]
    x_points[i,1]=x[1]
    w_points[i]=wgt*integrand([x[0],x[1]])
    #print(wgt)
    i+=1
    
print(x_points)
print(w_points)
```
