```python
import lhapdf
import numpy as np
import matplotlib.pyplot as plt

pdf_set_name = "MSHT20nnlo_as118"
pdf = lhapdf.mkPDF(pdf_set_name, 0)

xs = np.logspace(-4, 0, 1000)  
Q_values = [10, 100, 1000, 10000]  

flavors = {
    "u (up)": 2,
    "d (down)": 1,
    "s (strange)": 3,
    "c (charm)": 4,
    "b (bottom)": 5,
    "g (gluon)": 21
}

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for i, Q in enumerate(Q_values):
    ax = axs[i // 2, i % 2]  
    
    for flavor, pid in flavors.items():
        pdf_values = [pdf.xfxQ(pid, x, Q)  for x in xs]  
        ax.plot(xs, pdf_values, label=flavor)
    
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.set_xlabel(r"$x$")
    ax.set_ylabel(r"$x f(x, Q)$")
    ax.set_title(f"PDFs from {pdf_set_name} at Q = {Q} GeV")
    ax.legend()
    ax.grid(True)

plt.tight_layout()

plt.show()

dir(pdf)

```
